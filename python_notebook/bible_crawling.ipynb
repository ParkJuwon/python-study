```python
from bs4 import BeautifulSoup
import requests
import re
import time
```

```python
start_link = 'http://kcm.co.kr/bible/korea.html'
detail_link = 'http://kcm.co.kr{}'

resp = requests.get(start_link)
resp.raise_for_status()
resp.encoding = None
soup = BeautifulSoup(resp.text)

bible_detail_link_list = [anchor['href'] for anchor in soup.select('body > font > table > tr:not(:first-child) > td:nth-child(2) > font > a')]

with open("bible_list.txt", "w") as f:
    
    for link in bible_detail_link_list:
        print('1 sec sleep.... current page link : {}'.format(link))
        time.sleep(1)

        detail_page = detail_link.format(link.split('#')[0])
        detail_resp = requests.get(detail_page)
        detail_resp.raise_for_status()
        detail_resp.encoding = None
        detail_soup = BeautifulSoup(detail_resp.text)

        text_data_list = []
        
        for tag in detail_soup.select('body > form > font > span > table > tr > td:nth-child(2)'):
            if tag.select('font'):
                for remove_tag in tag.select('font'):
                    remove_tag.decompose()
            text_data_list.append(tag.text.replace('○', '').strip())
        
        middle_idx = int(len(text_data_list) / 2)
        first = ' '.join(text_data_list[:middle_idx])
        second = ' '.join(text_data_list[middle_idx:])
        
        f.write(first + '\n')
        f.write(second + '\n')
        
        
print('done')
```
