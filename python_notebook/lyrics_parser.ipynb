```python
from bs4 import BeautifulSoup
import requests
import re
import time
```

```python
# https://www.lyrics.co.kr/?p=609525

# 
# h1 page-header fa fa-align-justify
# <title>

# 
# div box-body font_size_up

url='https://www.lyrics.co.kr/?p={}'


music_list = []
with open("music_list.txt", "w") as f:
    for page_no in range(609592, 0, -1):
        if page_no % 100 == 0:
            print('1 sec sleep.... current page_no : {}'.format(page_no))
            time.sleep(1)

        resp = requests.get(url.format(page_no))
        soup = BeautifulSoup(resp.text)

        title = soup.find('title')
        lyrics = soup.select('body > div > div > section > div > div.col-md-8.col-sm-7 > div > div > div.box-body.font_size_up')

        if title and lyrics:
            music_name = title.get_text().split('|')[0].strip()
            music_name = re.sub(r'[\s]*[(][+][)][\s]*', '$$$', music_name)
            music_lyrics = lyrics[0].text.strip()
            music_lyrics = re.sub(r'\r\n', ' ', music_lyrics)
            music_lyrics = re.sub(r'\n', ' ', music_lyrics)
            music = music_name + '$$$' + music_lyrics
            f.write(music + '\n')

print('done')

#571177  
```

```python
lyrics_list = []

HANGUL_FIRST = ord('')
HANGUL_LAST = ord('')

with open("music_list.txt", "r") as f:
    while True:
        line = f.readline()
        if not line: break
            
        split_line = line.split('$$$')
        if len(split_line) == 3:
            lyrics = split_line[2]

            no_space_lyrics = lyrics.replace(' ', '')

            total_cnt = len(no_space_lyrics)
            hangul_cnt = sum([1 if HANGUL_FIRST <= ord(c) <= HANGUL_LAST else 0 for c in no_space_lyrics])
            ratio = float(hangul_cnt) / total_cnt * 100

            if hangul_cnt >= 100 and ratio >= 70:
                lyrics_list.append(lyrics)
            


with open("hangul_music_list.txt", "w") as file:
    for lyrics in lyrics_list:
            file.write(lyrics)
```
