```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 점 그리기

# 좌표 (4,3) 지정
a = np.array([4, 3])

# Figure 객체를 만든 후
fig = plt.figure()
# 1x1 개의 서브플롯(subplot)을 추가하고, 첫 번째 서브플롯을 사용
ax = fig.add_subplot(1, 1, 1)

# 변수 a의 값을 좌표로 사용하여 점을 그린다.
# a[0] 이 x좌표, a[1]이 y좌표, s는 점의 크기
ax.scatter(a[0], a[1], s=30)

# 점 옆에 문자 a를 표시, size는 문자의 크기
ax.text(a[0]+0.2, a[1]+0.2, 'a', size=15)

# x,y 축 눈금 범위를 -5 ~ 5로 지정
ax.set_xticks(range(-5, 6))
ax.set_yticks(range(-5, 6))

# 그리드를 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)

# x축과 y축 눈금의 비율을 1:1로
# 창의 크기를 변경해도 비율이 유지
ax.set_aspect('equal', adjustable='box')

# 다음 코드를 사용하여 원점에 좌표축이 보이게 함
# 왼쪽 y축과 아래 x축을 (0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# 위쪽과 오른쪽 축 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 그려진 창 보여주기
plt.show()
```

```python
# 화살표 그리기

# quiver함수 를 사용하여 2차원 평면에 화살표를 그리는 예제. 벡터를 나타내기 위해 사용
a = np.array([4, 3])

# Figure 객체를 만든 후
fig = plt.figure()
# 1x1 개의 서브플롯(subplot)을 추가하고, 첫 번째 서브플롯을 사용
ax = fig.add_subplot(1, 1, 1)

# 시작 위치는 0,0 이고 ,화살표가 그려지는 끝 위치 변수를 a 값으로
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)

# x축 y축 눈금 범위를 0에서 6으로 지정
ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
# 2 개의 서브 플롯 그리기

A = np.array([1, 2])
B = np.array([4, 3])

# 1개의 행과 2개의 열을 가지는 서브플롯 생성
f, ax = plt.subplots(1, 2)

# 다음과 같이 인덱스를 사용하여 각 서브플롯을 사용
# 각 서브플롯 상단에 텍스트를 추가
ax[0].title.set_text('A')
ax[1].title.set_text('B')

# 각 서브플롯에 시작점이 (0, 0)인 벡터를 그립니다.
ax[0].quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1)

# 서브플롯별로 그리드를 생성
start_x = -1
end_x = 5
start_y = -3
end_y = 4

for i in range(2):
    ax[i].axis([start_x, end_x, start_y, end_y])
    ax[i].set_xticks(range(start_x, end_x))
    ax[i].set_yticks(range(start_y, end_y))
    ax[i].grid(True)
    ax[i].set_aspect('equal', adjustable='box')
    
plt.show()
```

# 벡터의 정의

- 벡터는 크기와 방향으로 양을 나타내는 방법
- 스칼라는 크기만을 사용하여 양을 나타냄




- 집합의 경우에는 순서를 따지지 않기 때문에 같은 원소로만 구성되면 같다고 함
- 벡터의 경우는 구성하는 원소의 순서를 따지기 때문에 동일한 원소를 가진 벡터라도 원소의 순서가 같지 않으면 다른 벡터


$
\begin{bmatrix}1 \\ 2 \\5\end{bmatrix} = \begin{bmatrix}1 \\ 2 \\5\end{bmatrix} \neq \begin{bmatrix}2 \\ 1 \\5\end{bmatrix}
$


- 크기 n 인 모든 벡터의 집합을 $R^n$ 이라고 하며 n 차원을 가진 공간을 의미
- $R^2$ 는 크기 2인 벡터 (2차원 벡터) 의 집합으로 2차원 평면을 의미 
- 하나의 2차원 벡터는 2차원 평면인 $R^2$ 상의 한 점이다.
- $R^3$ 은 3차원 공간이며 3차원 벡터는 3차원 공간인 $R^3$상의 한 점

- 벡터는 화살표로 나타낼 수 있다.
- 예를들어 벡터 a의 경우 원점에서 시작하여 x좌표 4, y좌표 3인 점에 도달하는 화살표로 표현

$a = 
\begin{bmatrix}
4 \\ 3
\end{bmatrix}
$



```python
# 벡터 [4 3]
a = np.array([4,3])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# 벡터를 화살표로 나타냅니다
# 시작점은 원점, 끝점은 벡터 [4 3] 입니다.
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)

ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

- 벡터의 형태에 따라 '열벡터 (column vector)' 와 '행벡터(row vector)' 가 있다.
- 예를 들어, 벡터 a는 2개의 행과 1개의 열을 가지며, 크기 2인 열벡터
- 벡터 b는 1개의 행과 3개의 열을 가지며 크기 3인 행벡터
- 벡터는 성분으로 구성
- 예를들어, 벡터 a는 2개의 성분으로 구성, 첫번쨰 성분 $a_1$은 4, 두번째 성분 $a_2$는 3입니다.
$$
a = \begin{bmatrix}
4 \\ 3
\end{bmatrix}
, b = \begin{bmatrix} 2 & 5 & 1 \end{bmatrix}
$$

- 선형대수에서 기본적으로 벡터를 열벡터로 표현하며 필요시 전치(transpose) 연산을 하여 행벡터로 표현
- 전치는 벡터의 행과 열을 바꾸는 연산
$$
a^T = \begin{bmatrix} 4 & 3 \end{bmatrix}
$$

- 벡터를 화살표로 나타낼 때 화살표의 시작점이 어디이든 화살표의 크기와 방향만 같으면 동일한 벡터

```python
# 벡터 [4, 3]
a = np.array([4,3])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# 원점에서 시작하는 벡터a를 그린다
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)

# (0, -3) 에서 시작하는 벡터 a를 그린다
ax.quiver(0, -3, a[0], a[1], angles='xy', scale_units='xy', scale=1)

# (1, 2) 에서 시작하는 벡터 a
ax.quiver(1, 2, a[0], a[1], angles='xy', scale_units='xy', scale=1)

# (-3, 1)
ax.quiver(-3, 1, a[0], a[1], angles='xy', scale_units='xy', scale=1)

start = -5
end = 10

ax.set_xticks(range(start, end))
ax.set_yticks(range(start, end))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()

```

```python
# 넘파이에서 1차원 배열은 열벡터 또는 행벡터로 사용할 수 있기 때문에 1차원 넘파이 배열의 shape 속성에는 두번째 항이 없다.
# 전치 연산을 하더라도 shape에 변화가 없다

a = np.array([4, 3])
print(a.shape)
print(a.T.shape)
```

```python
# 1차원 넘파이 배열에 shape 속성에 크기를 지정하여, 엄격히 열벡터 또는 행벡터로 지정할 수 있다.
# 이 경우, 넘파이 배열은 2차원 배열이 된다

a = np.array([4, 3])

# 1 차원 넘파이 배열의 shape 속성에 두번째 원소가 없다.
print(a.shape)

#1 차원 넘파이 배열을 열벡터로 지정하려면 shape 속성의 두번째 요소를 1을 지정
a.shape = (2,1)

print(a)

# 행백터
a.shape = (1,2)
print(a)
```

- 벡터의 첫 번째 벡터의 성분을 x축 방향으로 얼마만큼 이동했는지 나타냄
- 두번째 성분은 y축 방향으로 얼마나 이동했는지 나타냄
- 예를 들어 벡터 a는 양의 x축 방향으로 4, 양의 y축 방향으로 3만큼 이동한다
- 피타고라스 정리로 구할 수 있다
$$
\begin{vmatrix}a\end{vmatrix} = \sqrt{4^2 + 3^2} = 5
$$


- 모든 벡터는 크기가 1인 단위벡터를 사용하여 표현가능
- 벡터 i,j 를 기저벡터라고 한다
- 2차원 좌표계는 기저벡터 i,j 에 의해 구성
$$
a = \begin{bmatrix}4 \\ 3\end{bmatrix} = 4\begin{bmatrix}1 \\ 0\end{bmatrix} + 3\begin{bmatrix}0 \\ 1\end{bmatrix} = 4i + 3j
$$

```python
i = np.array([1, 0])
j = np.array([0, 1])

# 단위 벡터로 벡터 [4 3] 을 나타낼 수 있다
a = 4*i + 3*j

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], '4*i + 3*j', size=15)

ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

### 덧셈
- 두 벡터의 크기가 같을 때 덧셈 연산을 할 수 있다.
- 벡터의 덧셈은 두 벡터의 대응하는 성분끼리 더하여 계산이 이루어짐
$$
a + b = \begin{bmatrix}a_1 \\ a_2\end{bmatrix} + \begin{bmatrix}b_1 \\ b_2\end{bmatrix} = \begin{bmatrix}a_1 + b_1 \\ a_2 + b_2 \end{bmatrix}
$$

- 넘파이에서는 '+ 연산자와 add 함수' 를 사용하여 덧셈을 계산할 수 있다.
$$
a + b = \begin{bmatrix}4 \\ 1\end{bmatrix} + \begin{bmatrix}-2 \\ 3\end{bmatrix} = \begin{bmatrix}2 \\ 4 \end{bmatrix}
$$


- 벡터의 성분은 x축 성분끼리, y축 성분끼리 더하는 것을 의미
- 기하학 적으로 나타내면 벡터 b의 시작점을 벡터 a의 끝점으로 옮겼을 떄 벡터 a 의 시작점과 벡터 b의 끝점을 이은 선이 두벡터의 합

```python
a = np.array([4, 1])
b = np.array([-2, 3])

print(a + b)
print(np.add(a, b))
```

```python
sum_v = a+b

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)

ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0] + b[0], a[1] + b[1], 'b', size=15)

# 벡터의 합을 그리면 벡터 a의 시작점과 벡터 b의 끝점을 잇는 벡터가 그려짐
ax.quiver(0, 0, sum_v[0], sum_v[1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(sum_v[0] * 0.5 + 0.3, sum_v[1] * 0.5, 'a + b', size=15, color='blue')

ax.set_xticks(range(0, 7))
ax.set_yticks(range(0, 7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

### 벡터의 밸셈
- 대응하는 성분끼리 뺄셈을 하여 계산

$$
a - b = \begin{bmatrix}a_1 \\ a_2\end{bmatrix} - \begin{bmatrix}b_1 \\ b_2\end{bmatrix} = \begin{bmatrix}a_1 - b_1 \\ a_2 - b_2 \end{bmatrix}
$$

- 넘파이에서는 '- 연산자와 subtract 함수' 를 사용하여 뺄셈을 계산할 수 있다.
$$
a - b = \begin{bmatrix}4 \\ 1\end{bmatrix} - \begin{bmatrix}-2 \\ 3\end{bmatrix} = \begin{bmatrix}6 \\ -2 \end{bmatrix}
$$

```python
a = np.array([4, 1])
b = np.array([-2, 3])

print(a - b)
print(np.subtract(a, b))
```

```python
sub = a-b

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)

# 벡터 a 끝점에서 시작되는 반대 방향의 벡터 b를 구함
ax.quiver(a[0], a[1], -b[0], -b[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0] - b[0], a[1] - b[1], 'b', size=15)

# 벡터의 뻴셈을 그리면 벡터 a의 시작점과 벡터 b의 끝점을 잇는 벡터가 그려짐
ax.quiver(0, 0, sub[0], sub[1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(sub[0] * 0.5 + 0.3, sub[1] * 0.5, 'a - b', size=15, color='blue')

ax.set_xticks(range(-2, 8))
ax.set_yticks(range(-4, 5))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

### 스칼라 곱
- 벡터의 스칼라를 곱하는 것
$$
ca = c\begin{bmatrix}a_1 \\ a_2\end{bmatrix} = \begin{bmatrix}ca_1 \\ ca_2 \end{bmatrix}
$$

- 넘파이에서는 '* 연산자와 multiply 함수' 를 사용하여 스칼라곱 계산을할 수 있다.
$$
3a = 3\begin{bmatrix}4 \\ 1\end{bmatrix} = \begin{bmatrix}12 \\ 3 \end{bmatrix}
$$

- 벡터의 양(+) 스칼라를 곱하면 벡터의 방향은 변하지 않고 크기만 변한다
- 벡터가 곱해진 스칼라의 크기만큼 늘어나거나 줄어듬

```python
a = np.array([4, 1])

c = 3

print(a * c)
print(np.multiply(a, c))
```

```python
ca = np.multiply(c, a)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)

# ca 구함,  시작점을 (0,1)
ax.quiver(0, 1, ca[0], ca[1], angles='xy', scale_units='xy', scale=1,color='blue')
ax.text(ca[0], ca[1]+1, '3a', size=15, color='blue')

ax.set_xticks(range(-2, 15))
ax.set_yticks(range(-3, 9))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

### 내적
- 두 벡터의 내적(dot product)은 다음과 같이 정의되며 계산 결과는 스칼라
- 내적은 대응하는 성분끼리 곱해서 모두 더하는 것
$$
a \cdot b = \begin{bmatrix}a_1 \\ a_2 \\ ... \\ a_n \end{bmatrix} \cdot \begin{bmatrix}b_1 \\ b_2 \\ ... \\ b_n \end{bmatrix} = a_1b_1 + a_2b_2 + ... + a_nb_n
$$

- 넘파이에서는 내적 계산을 위해 dot 함수를 제공
$$
a \cdot b = \begin{bmatrix}1 \\ 2 \end{bmatrix} \cdot \begin{bmatrix}2 \\ 3 \end{bmatrix} = 1 * 2 + 2 * 3 = 8
$$

```python
a = np.array([1, 2])
b = np.array([2, 3])

print(a @ b)
print(np.dot(a, b))
```

- 벡터의 크기는 다음과 같이 정의
$$
\begin{vmatrix}\begin{vmatrix}a\end{vmatrix}\end{vmatrix} = \sqrt{a_1^2 + a_2^2 + ... + a_n^2}
$$

- 넘파이에서는 벡터의 크기를 계산하기 위해 norm 함수를 제공
$$
\begin{vmatrix}\begin{vmatrix}a\end{vmatrix}\end{vmatrix} = \sqrt{1^2 + 2^2} = \sqrt{5}
$$

```python
a = np.array([1, 2])

print(np.linalg.norm(a))

# norm 함수의 계산결과는 제곱근 5와 같다
print(np.sqrt(5))
```

- 벡터 자기 자신에 대한 내적의 제곱근은 벡터의 크기와 같음
$$
\begin{vmatrix}\begin{vmatrix}a\end{vmatrix}\end{vmatrix} = \sqrt{a \cdot a}
$$

$$
a \cdot a = \begin{bmatrix}a_1 \\ a_2 \\ ... \\ a_n \end{bmatrix} \cdot \begin{bmatrix}a_1 \\ a_2 \\ ... \\ a_n \end{bmatrix} = a_1^2 + a_2^2 + ... + a_n^2
$$

- 넘파이를 사용하여 두 값이 같은지 확인
$$
\sqrt{a \cdot b} = \sqrt{\begin{bmatrix}1 \\ 2 \end{bmatrix} \cdot \begin{bmatrix}1 \\ 2 \end{bmatrix}} = \sqrt{1 * 1 + 2 * 2} = \sqrt{5}
$$

$$
\begin{vmatrix}\begin{vmatrix}a\end{vmatrix}\end{vmatrix} = \sqrt{1^2 + 2^2} = \sqrt{5}
$$

```python
a = np.array([1, 2])

# 자기 자신에 대한 내적의 제곱근
print(np.sqrt(np.dot(a, a)))

# 벡터의 크기
print(np.linalg.norm(a))
```

- 내적 $a \cdot b$ 는 벡터 a를 벡터 b에 정사영하여 구해지는 벡터와 벡터 b를 곱한 것
$$
a \cdot b = \begin{vmatrix}a\end{vmatrix}\begin{vmatrix}b\end{vmatrix}\cos\theta = \begin{vmatrix}a\end{vmatrix}\cos\theta * \begin{vmatrix}b\end{vmatrix}
$$

- 내적을 사용하여 벡터 b에 정사영 되는 벡터 a의 길이는 다음과 같이 구할 수 있다
$$
proj_ba =  \frac{a \cdot b}{\begin{vmatrix}b\end{vmatrix}}
$$
$$
\frac{a \cdot b}{\begin{vmatrix}b\end{vmatrix}} = \frac{\begin{vmatrix}a\end{vmatrix}\cos\theta * \begin{vmatrix}b\end{vmatrix}}{\begin{vmatrix}b\end{vmatrix}} = \begin{vmatrix}a\end{vmatrix}\cos\theta
$$

- 다음과 같이 45도 각도를 이루고 있는 벡터 a와 벡터 b의 내적은 2가지 방법으로 구할수 있다.

$$
a \cdot b = \begin{bmatrix}3 \\ 0\end{bmatrix} \cdot \begin{bmatrix}3 \\ 3\end{bmatrix} = 3 * 3 + 0 * 3 = 9
$$
$$
a \cdot b = \begin{vmatrix}\begin{vmatrix}a\end{vmatrix}\end{vmatrix}\begin{vmatrix}\begin{vmatrix}b\end{vmatrix}\end{vmatrix}\cos45 = 3 * 3\sqrt{2} * \frac{\sqrt{2}}{2} = 9
$$

```python
a = np.array([3, 3])
b = np.array([3, 0])

# 첫 번쨰 방법
print(np.dot(a, b))

# 두 번쨰 방법
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
print(norm_a * norm_b * np.cos(45*np.pi/180))
```

### 외적
- 평행하지 않는 두 벡터 v,w를 '외적(cross product)' 하면 두 벡터의 수직인 벡터를 구할 수 있으며 행렬식을 사용하여 다음과 같이 계산
- $i, j, k$ 는 해당 축에 해당하는 기저벡터
$$
v \times w = \begin{vmatrix}i & j & k \\ v_1 & v_2 & v_3 \\ w_1 & w_2 & w_3 \end{vmatrix}
$$

- 행렬식을 다음과 같이 전개하여 벡터를 구함
$$
v \times w = \begin{vmatrix}v_2 & v_3 \\ w_2 & w_3 \end{vmatrix}i - \begin{vmatrix}v_1 & v_3 \\ w_1 & w_3\end{vmatrix}j + \begin{vmatrix}v_1 & v_2 \\ w_1 & w_2 \end{vmatrix}k
$$

- 넘파이에선 cross 함수를 사용하여 외적 계산
$$
v \times w = \begin{vmatrix}i & j & k \\ 5 & 3 & 0 \\ 1 & 2 & 0 \end{vmatrix} = \begin{vmatrix}3 & 0 \\ 2 & 0\end{vmatrix}i - \begin{vmatrix}5 & 0 \\ 1 & 0\end{vmatrix}j +\begin{vmatrix}5 & 3 \\ 1 & 2\end{vmatrix}k 
= 0i - 0j + (5*2 - 3*1) = \begin{bmatrix}0, 0, 7\end{bmatrix}
$$

```python
v = np.array([5, 3, 0])
w = np.array([1, 2, 0])

print(np.cross(v, w))
```

### 행렬의 정의
- 행렬 (matrix) 는 숫자를 직사각형 모양으로 배열해놓은 것
- 구성하는 숫자를 원소라고 부름
- '행의 개수 x 열의 개수' 로 행렬의 크기를 나타냄
- m x n 행렬
$$
A = \begin{bmatrix}a_{11} & a_{12} & ... & a_{1n}
    \\ a_{21} & a_{22} & ... & a_{2n}
    \\ ... & ... & ... & ...
    \\ a_{m1} & a_{m2} & ... & a_{mn}
\end{bmatrix}
$$

- 행렬 A 의 크기는 2 x 3이다
$$
A = \begin{bmatrix}
1 & 2& 3 \\ 4& 5 & 6
\end{bmatrix}
$$

- 벡터도 행렬이다
- 하나의 행으로 구성된 행렬 a를 행벡터
- 하나의 열로 구성된 b를 열벡터
$$
a = \begin{bmatrix}1 & 2 & 3\end{bmatrix}, b = \begin{bmatrix}1 \\ 2 \\ 3\end{bmatrix}
$$

- 행렬의 열의 개수와 행의 개수가 같은 경우 '정사각행렬(square matrix)'이라고 함
$$
B = \begin{bmatrix}1 & 2 \\ 3 & 4\end{bmatrix}
$$

- 넘파이에서 행렬을 위한 matrix 함수를 제공하지만, 대부분 array 함수로 충분함

```python
A = np.array([[1,2,3], [4,5,6]])
print(A)

B = np.matrix([[1,2,3], [4,5,6]])
print(B)

# shape 속성으로 행렬의 크기를 확인 할 수 있다
print(A.shape)
print(B.shape)
```

```python
# 두 행렬의 행의 수와 열의 수가 같을 떄 두 행렬의 크기가 같다고 함
# 두 행렬 A, B가 모두 같으려면 두 행렬의 크기가 같고 원소가 모두 같아야함

A = np.array([[1,2], [3,4]])
B = np.array([[1,2], [3,4]])
C = np.array([[1,2], [3,4]])

print(A == B)
print(A == C)

# all 메소드를 사용하여 두 행렬이 같은지 리턴 받을 수 있음
print((A == B).all())
print((A == C).all())
```

```python
# 열벡터 행벡터 구하는 방법
A = np.array([[1,2,3,4], [5,6,7,8]])

# 행렬 A를 구성하는 행벡터
print(A[0])
print(A[0,:])
print(A[1])
print(A[1,:])

# 열벡터
print(A[:,0])
print(A[:,1])
print(A[:,2])
print(A[:,3])
```

```python
# 덧셈
# 원소끼리 더함

A = np.array([[1,2], [3,4]])
B = np.array([[4,1], [5,2]])

print(A + B)
print(np.add(A, B))
```

```python
# 뺄셈
# 원소끼리 뺌

A = np.array([[1,2], [3,4]])
B = np.array([[4,1], [5,2]])

print(A - B)
print(np.subtract(A, B))
```

###  곱셈

- 행렬의 곱셈이 가능하려면 첫 번째 행렬의 열의 수와 두 번째 행렬의 행의 수가 같아야 함
- m x n 인 행렬과 n x p 인 행렬을 곱하면 크기 m x p 인 행렬이 만들어짐
$$
AB = \begin{vmatrix} a_{11} & a_{12} & ... & a_{1n} 
\\  a_{21} & a_{22} & ... & a_{2n} 
\\  a_{m1} & a_{m2} & ... & a_{mn} 
\end{vmatrix}
\begin{vmatrix} a_{11} & a_{12} & ... & a_{1p} 
\\  a_{21} & a_{22} & ... & a_{2p} 
\\  a_{n1} & a_{n2} & ... & a_{np} 
\end{vmatrix}
$$

- 넘파이에선 dot 함수 또는 배열의 dot 메소드를 사용하여 곱셈 연산을 할 수 있다.

```python
A = np.array([[1,2,3], [4,5,6]])
B = np.array([[4,1], [5,2], [1,2]])

print(A.dot(B))
print(np.dot(A,B))
```

#### 주의할 점은 넘파이에서 * 연산자는 대응하는 원소간의 곱으로 계산
- '*' 연산자와 multiply 함수는 같은 연산을 함

```python
A = np.array([[1,2], [3,4]])
B = np.array([[4,1], [5,4]])

print(A*B)
print(np.multiply(A,B))
```

### 행렬 곱셈은 다음 법칙을 만족
1. 교환 법칙이 성립하지 않는다. 두행렬 순서가 바뀌면 결과가 달라짐
$$AB \neq BA$$

2. 곱셈과 덧셈 연산에 대해 다음과 같이 분배법칙이 성립
$$
A(B+C) = AB+AC
$$
$$
(B+C)D = BD+CD
$$

3. 결합법칙이 성립. 연산이 2번 이상 연속되는 경우 앞쪽 연산을 먼저 계산하여 구한 값과 뒤쪽 연산을 먼저 계산하여 구한 값이 항상 같다는 의미
$$
(AB)C = A(BC)
$$

4. 스칼라 c와 행렬 A, B에 대해 다음 법칙이 성립. 행렬 곱셈과 스칼라곱이 연속되는 경우 어느 것을 먼저 연산하는가에 관계없이 결과가 같음
$$
c(AB) = (cA)B
(AB)c = A(Bc)
$$


### 전치
- 행렬을 구성하는 원소의 열 인덱스와 행 인덱스를 바꾸는 것을 전치라고 함
$$
X = \begin{bmatrix}x_{11} & x_{12}
\\ x_{21} & x_{22}
\\ x_{31} & x_{32}
\end{bmatrix} = 
\begin{bmatrix} 1 & 2
\\ 3 & 4
\\ 5 & 6
\end{bmatrix}
$$

- 행렬 X에 전치 연산을 적용하면 원소들은 행 인덱스와 열 인덱스 바꾼 위치로 이동
$$
X^T = \begin{bmatrix} x_{11} & x_{21} & x_{31}
\\ x_{12} & x_{22} & x_{32}
\end{bmatrix} = \begin{bmatrix}1 & 3 & 5 \\ 2 & 4 & 6\end{bmatrix}
$$

- 넘파이에서 전치 연산을 transpose 함수 또는 배열의 T메소드 사용

```python
X = np.array([[1,2], [3,4], [5,6]])
print(X.T)
print(np.transpose(X))
```

### 항등 행렬
- 대각선 원소만 1이고, 나머지는 0 인 행렬을 '항등행렬(identity matrix)' 라고 함
$$
\begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix}
$$

$$
\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1\end{bmatrix}
$$

- 어떤 행렬에 항등 행렬을 곱해도 원래 행렬이 됨
$$
AI = IA = A
$$

- 넘파이에선 eye 함수를 사용하여 항등 행렬을 만들수 있다
- 항등행렬에서 특정 행렬을 곱하면 그대로 해당 행렬이 결과인 것도 확인 가능

```python
I = np.eye(3)
print(I)

# 배열 A 생성
A = np.arange(9).reshape(3,3)
print(A)

print((np.dot(A, I) == A).all())
print((np.dot(I, A) == A).all())
```

### 역행렬
- 행렬 A, B에 대해 다음식을 만족하면 행렬 B를 행렬 A의 역행렬 이라 하고 $A^{-1}$로 표기
- 여기에선 행의 개수와 열의 개수가 같은 정사각행렬에 대해서만 고려
$$
AB = BA = I
$$

- 크기가 2 x 2 행렬 $A = \begin{bmatrix}a & b \\ c & d \end{bmatrix}$ 의  경우 다음 식을 사용하여 역행렬을 구할 수 있다
- 이때 판별식이라고 부르는 ad-bc의 계싼 결과가 0 이 아닌 경우에만 역행렬을 구할 수 있다.
- 역행렬을 구할수 없는 행렬을 '특이행렬(Singular matrix)'라고 부름
$$
A^{-1} = \frac{1}{ad-bc}\begin{bmatrix}d & -b \\ -c & a\end{bmatrix}
$$

- 크기가 3x3 이상 행렬의 경우 가우스 소거법(gaussian elimination) 을 사용하여 역행렬을 구함
- 넘파이에선 inv 함수를 사용하여 역행렬을 구함

```python
A = np.array([[1,2], [3,4]])

# inv 함수를 사용하여 행렬 A의 역행렬을 구함
invA = np.linalg.inv(A)
print(invA)

print(np.dot(A,invA))

print(np.round(np.dot(A,invA)))
print(np.round(np.dot(invA, A)))
```

### 선형 결합
- n개의 백터 $x_1,x_2,...,x_n$ 와 n개의 스칼라 $c_1,c_2,...,c_n$ 가 있을 때 다음식을 선형결합(Linear Combination)이라고 함
- 벡터에 스칼라를 곱한다음 모두 더한것
$$
c_1x_1 + c_2x_2 + ... + c_nx_n
$$

- 같은 크기의 두 벡터를 선형 결합한 결과는 같은 크기의 벡터가 됨
$$
3x_1 + 5x_2 = 3\begin{bmatrix}1 \\ 2\end{bmatrix} + 5\begin{bmatrix}2 \\ 5\end{bmatrix} = \begin{bmatrix}13 \\ 31\end{bmatrix}
$$


#### 선형 독립과 선형 종속
- 스칼라 $c_1,c_2,...,c_n$ 가 모두 0이 아닐 때, 아래 식을 만족하면  백터 $x_1,x_2,...,x_n$ 는 '선형 종속(linearly dependent)' 라고 함
- 적어도 하나의 벡터를 다른 벡터의 조합으로 나타낼 수 있다
$$
c_1x_1 + c_2x_2 + ... + c_nx_n = 0
$$

- 계수 $c_1,c_2,...,c_n$ 가 모두 0일 떄만 위 식이 만족하면 벡터 $x_1,x_2,...,x_n$ 는 선형 독립 이라고 함
- 하나의 벡터를 다른 벡터의 조합으로 나타낼 수 없다는 의미

### 예제

3차원 공간에서 벡터 a,b 가 2차원 평면 $R^2$를 생성할 떄, 
세 번째 벡터 c가 벡터 a, b가 생성하는 평면위에 있다면 
벡터 a,b,c는 선형 종속이며 3차원 공간 $R^3$을 생성할 수 없다


```python
a = np.array([1, 5, 0])
b = np.array([4, 4, 0])
c = np.array([4, 7, 0])

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')


# 벡터 a
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='black', arrow_length_ratio=0.1)
ax.text(a[0], a[1], a[2], 'a', size=15)

# 벡터 b
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='black', arrow_length_ratio=0.1)
ax.text( b[0], b[1], b[2], 'b', size=15)

# 벡터 c
ax.quiver(0, 0, 0, c[0], c[1], c[2], color='blue', arrow_length_ratio=0.1)
ax.text( c[0], c[1], c[2], 'c', size=15,  color='blue')


ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_zlim(0, 8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20, azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

plt.show()
```

- 벡터 c는 벡터 a, b의 선형 결합으로 나타낼수 있기 떄문에 선형 종속이다
$$
0.75a + 0.81b = c
$$

- 벡터 c를 왼쪽으로 옮겨보면 0 이 아닌 계수를 가지는 벡터의 선형 결합이 된다.
- 3개의 개수가 0 일 때 외에도 다음과 같이 선형 결합식이 성립하므로 선형 종속이다
$$
0.75a + 0.81b - c = 0
$$

```python
a = np.array([1, 5, 0])
b = np.array([4, 4, 0])
c = np.array([4, 7, 0])

a = a * 0.75
b = b * 0.81

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')

ax.quiver(0, 0, 0, a[0], a[1], a[2], color='black', arrow_length_ratio=0.1)
ax.text(a[0], a[1], a[2], 'a', size=15)

ax.quiver(a[0], a[1], a[2], b[0], b[1], b[2], color='black', arrow_length_ratio=0.1)
ax.text( a[0]+b[0], a[1]+b[1], a[2]+b[2], 'b', size=15)

ax.quiver(0, 0, 0, c[0], c[1], c[2], color='blue', arrow_length_ratio=0.1)
ax.text( c[0], c[1], c[2], 'c', size=15,  color='blue')


ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_zlim(0, 8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20, azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')

plt.show()
```

### 예제

- 따라서 세 벡터는 선형 독립

### 랭크
- 행렬을 구성하는 열벡터 중 선형 독립인 열벡터의 개수를 '열랭크(column rank)'
- 행렬을 구성하는 행벡터 중 선형 독립인 행벡터의 개수를 '행랭크(row rank)'
- 행렬에 대한 열랭크와 행랭크는 항상 같으므로 구별없이 '랭크(rank)'라고 부름

- 다음 행렬 A의 랭크는 2이다
$$
A = \begin{bmatrix}0 & 1 & 2 \\ 1 & 2 & 1 \\ 2 & 7 & 8 \end{bmatrix}
$$
- 랭크 2라는 의미는 행렬을 구성하는 행벡터 중 2개는 선형 독립.
- 행렬을 구성하는 열벡터 중 2개는 선형 독립
- 2개의 행벡터 또는 열벡터를 결합하여 나머지 행벡터 또는 열벡터를 나타낼 수 있다.

- 예를 들어, 첫 번쨰 행벡터와 두번째 행벡터를 선형 결합하면 세 번째 행벡터를 얻을 수 있다.
$$
3\begin{bmatrix}0 & 1 & 2\end{bmatrix} + 2\begin{bmatrix}1 & 2 & 1\end{bmatrix} = \begin{bmatrix}2 & 7 & 8\end{bmatrix}
$$
- 두번째 열벡터와 세번째 열벡터를 선형 결합하면 첫번째 열벡터를 얻을 수 있다
$$
\frac{2}{3}\begin{bmatrix}1\\2\\7\end{bmatrix} + \frac{1}{3}\begin{bmatrix}2\\1\\8\end{bmatrix} = \begin{bmatrix}0\\1\\2\end{bmatrix}
$$

- 넘파이에서 rank 계산을 위해 matrix_rank 함수 제공

```python
A = np.array([[0, 1, 2], [1, 2, 1], [2, 7 ,8]])
np.linalg.matrix_rank(A)
```
