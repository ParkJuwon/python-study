```python
import pandas as pd
import numpy as np
```

### group by
- 아래의 세 단계를 적용하여 데이터를 그룹화(groupping) (SQL의 group by 와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할
    - operation 적용
    - 데이터 병합

```python
df = pd.read_csv('./train.csv')
df.head()
```

### GroupBy group 속성
- 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = df.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

### grouping 함수
- 그룹 데이터에 적용 가능한 통계 함수(NaN 은 제외하여 연산)
- count - 데이터 개수
- sum - 데이터의 합
- mean, std, var - 평균, 표준편차, 분산
- min, max - 최소, 최대값

```python
class_group.count()
```

```python
class_group.sum()
```

```python
class_group.mean()
```

```python
class_group.min()
```

```python
class_group.max()
```

```python
# 성별에 따른 생존률
df.groupby('Sex').mean()['Survived']
```

### 복수 column으로 groupping 하기
- groupby 에 column 리스트 전달
- 통계 함수를 적용한 결과는 multiindex 갖는 dataframe

```python
# 클래스와 성별에 따른 생존률
df.groupby(['Pclass', 'Sex']).mean()
```

```python
df.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')]
```

```python
df.groupby(['Pclass', 'Sex']).mean()['Survived']
```

### index를 이용한 group by
- index가 있는 경우, groupby 함수에 level 사용가능
    - level은 index의 depth를 의미하며 ,가장 왼쪽부터 0 부터 증가
- set_index 함수
    - column 데이터를 index 레벨로 변경
    
- reset_index 함수
    - 인덱스 초기화

```python
df.head()
```

```python
df.set_index('Pclass')
```

```python
df.set_index(['Pclass', 'Sex'])
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()
```

```python
df.set_index('Age').groupby(level=0).mean()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=0).mean()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=1).mean()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).mean()
```

```python
import math
def age_categorise(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
df.set_index('Age').groupby(age_categorise).mean()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).mean()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).aggregate([np.mean, np.sum, np.max])
```

### transform 함수
- groupby 후 transform 함수를 사용하면 원래의 index를 유지한 상태로 통계 함수를 적용
- 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산
- 따라서 새로 생성된 데이터를 원본 dataframe 과 합치기 쉬움

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass').transform(np.mean)
```

```python
df.groupby(['Pclass','Sex']).mean()
```

```python
df.groupby(['Pclass','Sex']).transform(np.mean)['Age']
```

```python
df['Age3'] = df.groupby(['Pclass','Sex']).transform(np.mean)['Age']
df
```

### pivot
- dataframe 형태를 변경
- 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시

```python
df.pivot('지역', '요일')
```

```python
df.pivot('요일', '지역')
```

```python
df.pivot('요일', '지역', '강수량')
```

### pivot_table
- 기능적으로 pivot과 동일
- pivot과 차이점
    - 중복되는 모호한 값이 있을경우 aggregation 함수 사용하여 값을 채움

```python
pd.pivot_table(df2, index='요일', columns='지역', aggfunc=np.mean)
```

### stack & unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경
    - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
- unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
    - stack 의 반대 operation
- 둘은 역의 관계

```python
new_df = df.set_index(['지역', '요일'])
new_df
```

```python
new_df.index
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

```python
new_df.stack(0)
```
