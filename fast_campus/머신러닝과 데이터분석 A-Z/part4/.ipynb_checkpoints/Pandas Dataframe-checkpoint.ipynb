### NaN 값 대체하기
- 평균으로 대체하기
- 생존자 / 사망자 별 평균으로 대체하기

```python
t_data['Age'].fillna(t_data['Age'].mean())
```

```python
# 생존자 나이 평균
mean1 = t_data[t_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean0 = t_data[t_data['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
t_data[t_data['Survived'] == 1]['Age'].fillna(mean1)
t_data[t_data['Survived'] == 0]['Age'].fillna(mean0)
```

```python
t_data.loc[t_data['Survived'] == 1, 'Age'] = t_data[t_data['Survived'] == 1]['Age'].fillna(mean1)
t_data.loc[t_data['Survived'] == 0, 'Age'] = t_data[t_data['Survived'] == 0]['Age'].fillna(mean0)
```

### Age 변수 변환하기
- 변환 로직을 만든 후, apply 함수로 적용

```python
import math
```

```python
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data['Age'].apply(age_categorize)
```

# 범주형 데이터 전처리 하기 (one-hot encoding)

### One-hot encoding
- 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요
- 범주형 데이터의 각 범주(category)를 column 레벨로 변경
- 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
- pandas.get_dummies 함수 이용
    - drop_first : 첫번쨰 카테고리값은 사용하지 않음

```python
train_data = pd.read_csv('./train.csv')
train_data.head()
```

```python
pd.get_dummies(train_data)
```

```python
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'])
```

```python
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True) # 첫번째 변수를 모르더라도 유추가 가능
```
